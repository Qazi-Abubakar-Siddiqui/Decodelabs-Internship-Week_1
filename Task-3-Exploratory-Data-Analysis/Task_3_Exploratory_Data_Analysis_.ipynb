```python
# =====================================================================
# STEP 1: AUTOMATED DATASET DOWNLOAD & LOAD CLEANED DATA
# =====================================================================
import os
import urllib.request
import pandas as pd
import numpy as np

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
filename = "winequality-red.csv"

if not os.path.exists(filename):
    urllib.request.urlretrieve(url, filename)

df = pd.read_csv(filename, delimiter=';')
df.drop_duplicates(inplace=True)  # Using the cleaned pipeline logic
df.columns = df.columns.str.replace(' ', '_')

# =====================================================================
# STEP 2: STATISTICAL ANALYSIS & INSIGHT SUMMARY
# =====================================================================
print("--- 1. Basic Statistical Description ---")
print(df.describe())

print("\n--- 2. Distribution of Target Feature (Quality Score) ---")
print(df['quality'].value_counts().sort_index())

print("\n--- 3. Correlation with Target (Quality) ---")
correlations = df.corr()['quality'].sort_values(ascending=False)
print(correlations)

print("\n--- 4. Outlier Analysis Using IQR (e.g., Residual Sugar) ---")
Q1 = df['residual_sugar'].quantile(0.25)
Q3 = df['residual_sugar'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['residual_sugar'] < (Q1 - 1.5 * IQR)) | (df['residual_sugar'] > (Q3 + 1.5 * IQR))]
print(f"Number of outlier rows detected in 'residual_sugar': {len(outliers)}")

print("\n--- EDA Summary Findings ---")
print(f"1. Wine quality scores range mostly from {df['quality'].min()} to {df['quality'].max()}.")
print(f"2. '{correlations.index[1]}' has the highest positive correlation with wine quality.")
print(f"3. '{correlations.index[-1]}' shows the strongest negative correlation with quality evaluation.")
```
