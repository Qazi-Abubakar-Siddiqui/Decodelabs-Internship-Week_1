```python
# ==========================================
# SETUP & DEPENDENCIES
# ==========================================
# First, ensure you have the required libraries installed:
# !pip install pandas numpy requests matplotlib seaborn scikit-learn

import os
import urllib.request
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ML Libraries for Task 5
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Set visualization style
sns.set_theme(style="whitegrid")
```

```python
print("--- TASK 1: DATA COLLECTION & UNDERSTANDING ---")

# URL of the authentic UCI dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
dataset_path = "winequality-red.csv"

# Download the file if it doesn't exist locally
if not os.path.exists(dataset_path):
    print("Downloading dataset from UCI Repository...")
    urllib.request.urlretrieve(url, dataset_path)
    print("Download complete!")
else:
    print("Dataset already exists locally.")

# Load the dataset (Note: UCI uses semicolons ';' as delimiters for this file)
df = pd.read_csv(dataset_path, sep=";")
print("Dataset loaded successfully!\n")
```

```python
# Understand dataset size
rows, cols = df.shape
print(f"Dataset Size: {rows} rows and {cols} columns.\n")

# Identify columns and data types
print("Columns and Data Types:")
print(df.info())

# Preview the first few rows
print("\nFirst 5 rows of the dataset:")
print(df.head())
```
