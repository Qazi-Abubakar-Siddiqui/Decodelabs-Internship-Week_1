```python
# =====================================================================
# STEP 1: DATASET SETUP
# =====================================================================
import os
import numpy as np
import urllib.request
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
filename = "winequality-red.csv"

if not os.path.exists(filename):
    urllib.request.urlretrieve(url, filename)

df = pd.read_csv(filename, delimiter=';')
df.drop_duplicates(inplace=True)
df.columns = df.columns.str.replace(' ', '_')

# Set aesthetic styling
sns.set_theme(style="whitegrid")

# =====================================================================
# STEP 2: CHART CREATION & STORYTELLING
# =====================================================================

# Chart 1: Distribution of Wine Quality Scores
plt.figure(figsize=(7, 5))
sns.countplot(x='quality', data=df, palette='dark:salmon_r')
plt.title('Distribution of Red Wine Quality Scores', fontsize=14, weight='bold')
plt.xlabel('Quality Score', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()

# Chart 2: Feature Correlation Heatmap
plt.figure(figsize=(10, 8))
mask = np.triu(np.ones_like(df.corr(), dtype=bool)) # Hide upper triangle for cleaner look
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='coolwarm', mask=mask, cbar_kws={"shrink": .8})
plt.title('Correlation Matrix of Wine Characteristics', fontsize=14, weight='bold')
plt.show()

# Chart 3: Impact of Alcohol Content on Quality
plt.figure(figsize=(8, 5))
sns.boxplot(x='quality', y='alcohol', data=df, palette='vlag')
plt.title('Alcohol Content Variation Across Quality Metrics', fontsize=14, weight='bold')
plt.xlabel('Quality Score', fontsize=12)
plt.ylabel('Alcohol (%)', fontsize=12)
plt.show()

print("\n--- Key Visual Insights Revealed ---")
print("• The countplot demonstrates that the quality scale is highly unbalanced, primarily centering around scores 5 and 6.")
print("• The boxplot explicitly illustrates that higher-rated wines (scores 7 and 8) systematically carry a noticeably higher alcohol content.")
```
